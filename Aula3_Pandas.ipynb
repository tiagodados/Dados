````python
# Importando a biblioteca 

import pandas as pd
````

````python
# Lendo dados tabulares

credit = pd.read_csv("/credit_card_clients.csv")
````

````python
# Acessando os rótulos das linhas

credit
````

````python
# Acessando os rótulos das colunas

credit.columns 
````

````python
credit.keys()
````

- **Observação**

`A operação keys() retorna o mesmo objeto que a propriedade columns.`

````python
# Identificando os valores do dataset a partir de um objeto do tipo ndarray

credit.values
````

````python
credit.to_numpy()
````

- **Observação**

`Tanto a operação to_numpy() quanto o atributo values retorna uma representação numpy.ndarray dos valores contidos no DataFrame.`

````python
# Identificando os tipos de dados contidos em cada uma das colunas

credit.dtypes
````

- **Observação**

`A propriedade dtypes retorna uma série com os tipos de dados de cada coluna do DataFrame.`

````python
# Iterando linhas e colunas de um DataFrame

for index, row in credit.iterrows():
  print(index)
  print(row)
````

- **Observação**

`Laços do tipo For: através das estruturas de repetição podemos realizar uma computação até que uma certa condição seja satisfeita:`
