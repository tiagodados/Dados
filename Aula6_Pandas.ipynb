````python
# Importando a biblioteca 

import pandas as pd
````

````python
# Lendo dados tabulares

credit = pd.read_excel("/content/clientes_credito.xlsx")
````

````python
# Extraindo um sumário das estatísticas descritivas
# describe somente analisa colunas númericas

credit.describe()
````

O método `describe()` retorna estatísticas descritivas:

- count: Número de elementos da coluna com valores diferentes de NaN.

- mean: Média dos valores nas colunas.

- std: Desvio padrão dos valores nas colunas.

- min: Valor mínimo na coluna.

- max: Valor máximo na coluna.

- percentis: primeiro quartil (25% das observações abaixo e 75% acima), segundo quartil (mediana, deixa 50% das observações abaixo e 50% das observações acima) e terceiro quartil (75% das observações abaixo e 25% acima).

Posso também filtrar os cálculos estatísticos para uma coluna em específico:

````python
credit["valor_credito"].describe()
````

> **Variável aleatória (v.a.)**

Em um estudo, há sempre a necessidade de se focar em um ou mais atributos (características) dos elementos que compõem minha população.

*Na situação problema em questão o que vem a ser minha população?*

- Atributos quantitativos: `valor_credito`

- Atributos qualitativos (categóricos): `status_pag_set_2005`

**Definição v.a.**: é uma função que associa cada elemento da população a um número:

Captura de tela de 2022-09-05 10-12-31.png


Na prática o que é uma variável aleatória? Voltemos ao Dataset `credit`:

````python
credit['idade']
````

No exemplo acima minha **v.a.** é definida por:

`X: {24, 26, 34, 37, 57, ..., 39, 43, 37, 41, 46}`

> **Caracterização de uma v.a.:**

Aspectos principais:
- Posição (tendência central).
- Dispersão.
- Forma.

Vamos então entender o que são as medidas de tendência central:

````python
# Média

credit["valor_credito"].mean()
````

````python
# Mediana

credit["valor_credito"].median()
````

````python
# Moda

credit["valor_credito"].mode()
````

- **MÉDIA**: ponto de equilíbrio da distribuição.

Captura de tela de 2022-09-05 10-42-41.png

- **MEDIANA**: ponto que divide a distribuição em duas partes iguais (equiprováveis).

Captura de tela de 2022-09-05 10-40-23.png

- **MODA**: valor(es) que ocorre(m) com a maior frequência.

Captura de tela de 2022-09-05 10-44-46.png

Vamos agora entender o que são as medidas de dispersão:

````python
# Variância

credit["valor_credito"].var()
````

````python
# Desvio padrão
# variabilidade do dado em relação a média

credit["valor_credito"].std()
````

- **VARIÂNCIA**: média dos desvios quadráticos da v.a. (em relação à média).

- **DESVIO PADRÃO**: raiz quadrada da variância.
