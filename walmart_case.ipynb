```python
#Rolling Filters
df = pd.read_csv("/content/walmart_case.csv",
usecols=['Date','Weekly_Sales'], parse_dates=['Date'], index_col='Date')
df_monthly = df.resample('M').sum().sort_index()
df_monthly.head()
```

```python
@interact(window_size=range(1,36,1))
def rolling_mean(window_size):
  plt.figure(figsize=(15,7))
  df_monthly.Weekly_Sales.rolling(window_size).mean().plot()
  plt.show()
```
