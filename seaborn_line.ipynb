```python
# Importando as bibliotecas

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
```

```python
# Lendo dados tabulares

credit = pd.read_excel("/content/clientes_credito.xlsx")
```

```python
credit
```

```python
# Agrupando dados

credit_group = credit.groupby(['idade']).mean().reset_index()
credit_group
```

```python
# GRÁFICO COM A MATPLOTLIB

plt.figure(figsize=(10,8))

plt.title("Valor de Crédito concedido aos clientes segundo a idade")

plt.xlabel("Idade")
plt.ylabel("Média de crédito concedido")


plt.plot(credit_group["idade"], credit_group["valor_credito"]);
plt.grid(color='purple')
```

```python
# GRÁFICO COM A SEABORN

#Análise de extrato bancário

fig, axes = plt.subplots(2,2, figsize=(20,10))

sns.lineplot(data=credit, x="idade", y="extrato_jun_2005", hue="sexo", style="sexo", palette="Set2", ax=axes[0,0]).set(title="Extrato Junho de 2005", xlabel=None)
sns.lineplot(data=credit, x="idade", y="extrato_jul_2005", hue="sexo", style="sexo", palette="Set2", ax=axes[0,1]).set(title="Extrato Julho de 2005", xlabel=None)
sns.lineplot(data=credit, x="idade", y="extrato_ago_2005", hue="sexo", style="sexo", palette="Set2", ax=axes[1,0]).set(title="Extrato Agosto de 2005")
sns.lineplot(data=credit, x="idade", y="extrato_set_2005", hue="sexo", style="sexo", palette="Set2", ax=axes[1,1]).set(title="Extrato Setembro de 2005");



# 1=Masculino, 2=Feminino
```

```python
# GRÁFICO COM A PLOTLY

fig = go.Figure()

from plotly.subplots import make_subplots

fig = make_subplots(rows=3, cols=2,shared_yaxes=True,
                    subplot_titles=("Set", "Ago", "Jun", "Jul","Mai","Apr"))

fig.add_trace(go.Scatter(
    x=credit_group['idade'], 
    y=credit_group['valor_pag_set_2005'],
    line=dict(color='red', width=1),
    name='Valor pago 09/2005'), row=1, col=1)

fig.add_trace(go.Scatter(
    x=credit_group['idade'],
    y=credit_group['valor_pag_ago_2005'],
    line=dict(color='yellow', width=1),
    name='Valor pago 08/2005'
), row=1, col=2)

fig.add_trace(go.Scatter(
    x=credit_group['idade'],
    y=credit_group['valor_pag_jul_2005'],
    line=dict(color='green', width=1),
    name='Valor pago 07/2005'
), row=2, col=1)

fig.add_trace(go.Scatter(
    x=credit_group['idade'],
    y=credit_group['valor_pag_jun_2005'],
    line=dict(color='blue', width=1),
    name='Valor pago 06/2005'
), row=2, col=2)

fig.add_trace(go.Scatter(
    x=credit_group['idade'],
    y=credit_group['valor_pag_mai_2005'],
    line=dict(color='crimson', width=1),
    name='Valor pago 05/2005'
), row=3, col=1)

fig.add_trace(go.Scatter(
    x=credit_group['idade'],
    y=credit_group['valor_pag_abr_2005'],
    line=dict(color='purple', width=1),
    name='Valor pago 04/2005'
), row=3, col=2)

fig.update_layout(title_text="Valor pago em 2005 o longo dos meses de acordo com a idade dos clientes")

fig.update_yaxes(title_text="Idade", row=1, col=1, titlefont_size=10)
fig.update_yaxes(title_text="Idade", row=1, col=2, titlefont_size=10)
fig.update_yaxes(title_text="Idade", row=2, col=1, titlefont_size=10)
fig.update_yaxes(title_text="Idade", row=2, col=2, titlefont_size=10)
fig.update_yaxes(title_text="Idade", row=3, col=1, titlefont_size=10)
fig.update_yaxes(title_text="Idade", row=3, col=2, titlefont_size=10)
fig.update_xaxes(row=1, col=1, titlefont_size=10)
fig.update_xaxes(row=1, col=2, titlefont_size=10)
fig.update_xaxes(row=2, col=1, titlefont_size=10)
fig.update_xaxes(row=2, col=2, titlefont_size=10)
fig.update_xaxes(title_text="Valor pago (dólar NT)", row=3, col=1, titlefont_size=10)
fig.update_xaxes(title_text="Valor pago (dólar NT)", row=3, col=2, titlefont_size=10)

fig.show()

# Salvando o gráfico interativo
fig.write_html("valor_pago_idade.html")
```
