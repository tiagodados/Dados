````python
# Importando as bibliotecas

!pip install rasterio

import numpy as np
import rasterio
from rasterio.plot import show
````

````python
dataset = rasterio.open('/content/crop_rapideye.tif') # Lendo a imagem
````

````python
# Convertendo o dataset em um objeto do tipo ndarray

dataset_array = dataset.read(1)
````

````python
dataset_array
````

````python
dataset_array
show(dataset)
````

````python
# Calculando a média

dataset_array.mean() # ponto de equilibrio da distribuicao é a soma de todos os valores divididos pela quantidade de elementos
````

````python
# Calculando o desvio padrão

dataset_array.std() #desvio padrão da média
````

````python
# Identificando o valor mínimo

dataset_array.min() #valor minimo
````

````python
# Identificando o valor máximo

dataset_array.max() # valor máximo
````

````python
# Somando todos os valores da array numpy

dataset_array.sum() #soma total de valores brutos
````

````python
# Identificando os valores únicos
unique_values = np.unique(dataset_array)
unique_values
````
