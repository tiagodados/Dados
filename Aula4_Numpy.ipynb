````python
# Retorna uma tupla de inteiros que indica o número de elementos armazenados em cada dimensão da array

print('shape', dataset_array.shape)

# Número de axes - dimensões

print('ndim', dataset_array.ndim)

# Tipo de dado

print('dtype', dataset_array.dtype)

# Comprimento de um elemento da array, em bytes

print('itemsize', dataset_array.itemsize)

# Número total de elementos da array

print('size', dataset_array.size)
````
